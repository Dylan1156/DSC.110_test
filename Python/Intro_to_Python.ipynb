# **Functions**

```python
#Celsius to Fahrenheit conversion equation
temp = (4 * 9/5) + 32
print(temp)
```

```python
temp = (-6 * 9/5) + 32
print(temp)
```

```python
# function converts Celsius to Fahrenheit
# functions allow code reuse and organization
def c_to_f(c_temp):
  f_temp=(c_temp * (9/5)) + 32
  return f_temp
```

```python
f_temp = c_to_f(4)
print(f_temp)
```

Formatted string literals

```python
# f-string inserts the calculated value into the sentence
# makes output easier to read
print(f'The temperature now is {c_to_f(35)}F')
```
